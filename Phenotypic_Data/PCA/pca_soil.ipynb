# Dimension Reduction of soil data using PCA

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('Soilvalues.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
df_scaled = pd.DataFrame(scalar.fit_transform(df), columns=df.columns)
df_scaled
```

```python
pca = PCA()
df_pca = pd.DataFrame(pca.fit_transform(df_scaled))
df_pca
```

```python
import matplotlib.pyplot as plt
pd.DataFrame(pca.explained_variance_ratio_).plot.bar()
plt.legend('')
plt.xlabel('Principal Components')
plt.ylabel('Explained Varience')
```

## Export the scores

```python
# top_pc = df_pca[["0", "1", "2", "3", "4", "5", "6"]]
top_pc = df_pca.iloc[:, :6].copy()
top_pc
```

```python
top_pc.columns = ['SPC1', 'SPC2', 'SPC3', 'SPC4', 'SPC5', 'SPC6']
# df_pca.rename(columns={'0': 'Entry', '1': 'SPC1', '2': 'SPC2', '3': 'SPC3', '4': 'SPC4', '5': 'SPC5', '6': 'SPC6'}, inplace=True)
top_pc
```
