```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as patches
import matplotlib.transforms as transforms
```

```python
Data = pd.read_csv("Weather_Data_Stacked.csv")
print(Data)
```

```python

ax = sns.scatterplot(x="tmin", y="tmax", data=Data)
ax.set_title("Min and Max Temperature Correlation")
ax.set_ylabel("Maximum Temperature")
ax.set_xlabel("Minimum Temperature")
```

```python
sns.lmplot(x="tmin", y="tmax", hue="year", data=Data)
```

Correlations Matrix

```python
Corr_Data = Data[['tmax', 'tmin', 'prcp', 'GDD', 'temp', 'srad']]
```

```python

cormat = Corr_Data.corr()
round(cormat, 2)
sns.heatmap(cormat)
```
